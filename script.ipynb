```python
df_tracks.set_index("release_date", inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks["artists"].iloc[18]
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
df_tracks.duration.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_df = df_tracks.drop(['key','mode','explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin= -1, vmax= 1, center= 0, cmap= 'inferno', linewidths=2, linecolor='black')
heatmap.set_title('Correlation Heatmap Between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
len(sample_df)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', scatter_kws={'color': 'cyan'}, line_kws={'color': 'red'}).set(title = "Loudness vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', scatter_kws={'color': 'green'}, line_kws={'color': 'blue'}).set(title = "Popularity vs Acousticness Correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=4, kind='hist').set(title='Number of Songs Per Year')
```

```python
total_dr = df_tracks.duration
fig = plt.subplots(figsize = (18,7))
fig = sns.barplot(x = years, y = total_dr, errwidth=False).set(title='Year vs Duration')
plt.xticks(rotation=90)
```

```python
sns.set_style(style='whitegrid')
fig = plt.subplots(figsize=(10,5))
fig = sns.lineplot(x = years, y = total_dr).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
df_genre = pd.read_csv('./datasets/SpotifyFeatures.csv')
df_genre.head()
```

```python
fig = plt.subplots(figsize=(5,8))
sns.color_palette('rocket', as_cmap=True)
fig = sns.barplot(y = 'genre', x = 'duration_ms', data = df_genre).set(title='Duration of the Songs in Different Genres')
plt.xlabel('Duration in Milliseconds')
plt.ylabel('Genres')
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize=(10,5))
famous = df_genre.sort_values('popularity', ascending = False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = famous).set(title = "Top 5 Genres by Popularity")
```
